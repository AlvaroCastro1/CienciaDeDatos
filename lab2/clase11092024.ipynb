```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# vamos a definir los limites de plano
y1 = -1
y2 = 10
x1 = -1
x2 = 10

# Definir ejes
plt.axis([x1, x2, y1, y2])
# añadimos dos lineas que representen los ejes X y Y
plt.axhline(0, color='black',linewidth=1) # Eje X
plt.axvline(0, color='black',linewidth=1) # Eje Y 

# se hace visible una cuadricula
plt.axis('on')
plt.grid(True)

plt.title("Vectores Actividad clase 11/09/2024")

#  añadimos el vector A
plt.arrow(0, 0, 3, 4, head_length=0.5, head_width=0.3, color="r")
# la magnitud del vector
plt.text(1, 0.5, "A = 5", fontsize=8, style="italic", rotation=45)
# sus proyecciones
plt.text(3, 0, "longitud = 3", fontsize=8, rotation=90)
plt.text(0, 4, "longitud = 4", fontsize=8)
```
