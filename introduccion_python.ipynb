```python
# Ejercicio
"""
Agregar un numero aletorio de valores 4 a una lista
y luego el valor 1.5
1) mostrar la lista completa
2) Eliminar el valor 4 y mostrar la lista despues de cada eliminacion
Nota: el codigo no debe lanzar excepciones
"""

import numpy as np
r = np.random.randint(1, 11) # no llega a 11
r
```

```python
lista = list()
for i in range(r):
  lista.append(4)
print(f"N: {len(lista)} \nlista:{lista}")
```

```python
# agregar el 1.5
lista.append(1.5)
lista
```

```python
# Eliminar el valor 4 y mostrar la lista despues de cada eliminacion
# while True:
while (4 in lista):
  lista.remove(4)
  print(lista)
```

```python
while True:
  try:
    lista.remove(4)
  except:
    break
    print("No esta")
  print(lista)
```
