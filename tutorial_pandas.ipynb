### Primero se a√±aden las librerias

```python
# librerias que se usan
import pandas as pd
#  libreria para usar archivo desde drive
from google.colab import drive
# para generar numeros aleatorios
import random
```

Un **dataframe** tiene **filas** y **columnas**. Pero estas solo son de un tipo.
Y podemos leer desde un archivo...

```python
# vamos a cargar los datos del drive
drive.mount('/content/drive')

# ahora leemos los datos y manejamos el index como la columna id
df = pd.read_csv('/content/drive/MyDrive/2024Cienciadatos/dataset.csv', index_col="id")

# aqui solo mostramos los ultimos 3 registros
df.tail(3)
```

```python
# podemos ver las estadisticas descriptivas para variables numericas
df.describe()
```

# Limpieza de datos

```python
# vamos a filtrar el df para que no haya datos NaN
df.dropna()
```

```python
# o podemos rellenar los NaN
df.fillna(0)
```

```python
# o podemos especificar a cada columna como se debe rellenar

df = df.fillna({ "retweets": 0, "mentions": -1 })
```

# An√°lisis de datos

```python
# aqui solo se consultan los datos de una columna
df["favorites"]
```

```python
# o varias columnas con ayuda de una lista
df[ ["favorites", "full_text"]]
```

# filtrado mediante indices con iloc

```python
# un solo elemento
df.iloc[0]
```

```python
# multiples elementos
df.iloc[0:3]
```

```python
# con indices de fila
df.iloc[[0,2]]
```

```python
# con identificadores (elementos individuales en una lista)
df.loc[[183721	, 183722]]
```

```python
# se puede filtrar al mismo tiempo filas y columnas
df.loc[[183721	, 183722], ["favorites", "full_text"]]
```

```python
# tambien se puede filtrar por condiciones
df[df["favorites"] > 400]
```

```python
# tambien se pueden concatenar condiciones
df[(df["favorites"] > 400) & (df["mentions"] > 20)]
```

```python
# tambien para coincidencias con texto
df[df["full_text"].str.contains("Programming")]
```

```python
# ahora podemos crear nuevas columnas mediante datos que ya existen en el df
def calcularGanancias(retweets):
  return retweets * random.randint(3, 5)

# se agrega una columna al final
# por cada valor de la columna se aplica una funcion
df["ganancias"] = df["retweets"].apply(calcularGanancias)
```

```python
# aqui ya podemos observar la nueva columna
df.head()
```

```python
# ahora vamos a calcular una nueva columna
def popularidad(fila):
  return fila["followees"] / fila["followers"]

# se aplica a todas las columnas
df["popularidad"] = df.apply(popularidad, axis=1)
```
